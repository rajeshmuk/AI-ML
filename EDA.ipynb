```python
#8 Visualize how popularity of genres has changed over the years
# Objective here is to set the index as release year
trimmed_movie_details = movie_details
trimmed_movie_details = movie_details.drop(['movie id','release date','CountGenre'],axis=1)

trimmed_movie_details.set_index('rel year',inplace=True)
trimmed_movie_details
```

```python
#8 Visualize how popularity of genres has changed over the years
#Movie Releases per year
moviePerYear=pd.DataFrame()

for i in movielist:
    moviePerYear[i]=trimmed_movie_details[i].groupby('rel year').sum()
moviePerYear['total']=trimmed_movie_details.groupby('rel year')['movie title'].count()
moviePerYear.index
moviePerYear
```

```python
#8 Visualize how popularity of genres has changed over the years
#Movie Releases per year on percentage basis
moviePerYear.div(moviePerYear['total'], axis= 0) * 100
```

```python
#8 Visualize how popularity of genres has changed over the years
plotdata=moviePerYear
ax=plotdata.plot(figsize=(15,15))
ax.set_xlim(1990,1999)
```

```python
#8 Visualize how popularity of genres has changed over the years
#Using Heatmap to see the popularity of genres per Year
cor = moviePerYear[['Action', 
              'Adventure', 
              'Animation', 
              'Childrens', 
              'Comedy', 
              'Crime',
              'Documentary',
              'Drama',
              'Fantasy', 
              'Film-Noir',
              'Horror',
              'Musical',
              'Mystery',
              'Romance',
                    'Sci-Fi',
                    'Thriller',
                    'War',
                    'Western'
                   ]].corr()
cor
sns.heatmap(cor,annot=True,cmap='plasma',vmin=-1,vmax=1);
```

```python
#9. Find the top 25 movies in terms of average ratings for movies that have been rated more than 100 times 
# Find all the movie details which has been rated more than 100 times
merged_data=pd.merge(movie_rating[['movie id', 'rating']],movie_details[['movie id', 'movie title']],how='inner',on='movie id')

df=pd.DataFrame()
df['ratingcount']=merged_data.groupby(['movie id'])['rating'].count().sort_values(ascending=False)
df['average']=merged_data.groupby(['movie id'])['rating'].mean()
#df['movietitle']=merged_data['movie title']
df[df['ratingcount']>100]
```

```python
#9. Find the top 25 movies in terms of average ratings for movies that have been rated more than 100 times
# Merge the data and find the top 25 movies based on rating count and average rating
final_data = pd.merge(df,movie_details[['movie id', 'movie title']],how='inner',on='movie id')
final_data.sort_values(by=['ratingcount','average'],ascending=[False,False]).head(25)
```

```python
#9. Find the top 25 movies in terms of average ratings for movies that have been rated more than 100 times
#Movie Title vs Rating of the 25 data
final_data.head(25).plot(kind='bar',x='movie title',y='average',figsize=(15,15));
```
